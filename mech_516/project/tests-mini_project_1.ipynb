```python
def plot_test(riemann, W, W_target, t, i, axes):
    x0 = riemann.x0
    plot_riemann(riemann, np.linspace(0,t*1.2, 100), axes[0])
    axes[0].plot(x, np.ones_like(x)*t, 'g-.')
    axes[0].set_xlim(-0.5, 0.5)
    axes[0].set_title('Test %i' % i, size=24)
    titles = ['Density', 'Velocity', 'Pressure']
    for i in range(3):
        axes[i+1].plot(x, W_target[:,i], 'b-')
        axes[i+1].plot(x, W[:,i], 'r.', ms=2)
        axes[i+1].set_xlim(-0.5, 0.5)
        axes[i+1].set_title(titles[i])
    intersect = []
    for wave in [riemann.left_wave, riemann.right_wave]:
        try:
            intersect.append(wave.speed[0]*t + riemann.x0)
            intersect.append(wave.speed[1]*t + riemann.x0)
        except IndexError:
            intersect.append(wave.speed*t + riemann.x0)
    intersect.append(riemann.u_star*t + riemann.x0)
    for ax in axes:
        for point in intersect:
            ax.axvline(point, color='black', linestyle='-.')
```

```python
fig, axes = plt.subplots(nrows=4, ncols=3, sharex=True, squeeze=True, figsize=(30, 30))
axes = np.array(axes).T
axes1, axes2, axes3 = axes
plot_test(riemann1, W1, W1_target, t1, 1, axes1)
plot_test(riemann2, W2, W2_target, t2, 2, axes2)
plot_test(riemann3, W3, W3_target, t3, 3, axes3)
plt.tight_layout()
```
