```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import os
from euler import schemes
from riemann.riemann import Riemann
from riemann.plot import plot_riemann, plot_particles
from linear.plot_utils import format_axes, latexify
%matplotlib inline
```

```python
# latexify()
make_path = lambda x: os.path.abspath(os.path.join('project3_tex', 'figs', x + '.pdf'))
```

```python
# Initial conditions
rho_l = 1
u_l = 0
p_l = 2
rho_r = 1
u_r = 0 
p_r = 1
gamma = 1.4
t_end = 25
dx = 1
dt = 0.5
nx = 100
```

```python
# Generating arrays
x = np.arange(-49.5, 49.5 + dx, dx)
left = x < 0
right = x > 0
rho0 = np.ones_like(x)*rho_l
rho0[right] = rho_r
u0 = np.ones_like(x)*u_l
u0[right] = u_r
p0 = np.ones_like(x)*p_l
p0[right] = p_r
```

```python
# Primitive values
W_l = np.array([rho_l, u_l, p_l])
W_r = np.array([rho_r, u_r, p_r])
```

```python
# Find exact solution
riemann = Riemann(W_l, W_r, gamma)
riemann.update()
rho_ex, u_ex, p_ex = riemann.evaluate_states(x, t_end).T
```

```python
def max_eigenvalue(rho, u, p):
    c = np.sqrt(gamma*p/rho)
    # Calculate maximum of each eigenvalue
    y1max = np.max(np.abs(u - c))
    y2max = np.max(np.abs(u))
    y3max = np.max(np.abs(u + c))
    ymax = max((y1max, y2max, y3max))
    return ymax
```

```python
def integrate(limiter, rho0, u0, p0, CFL=0.8, verbose=False):
    rho, u, p = rho0.copy(), u0.copy(), p0.copy()
    t = 0
    while True:
        if t >= 25:
            return rho, u, p

        ymax = max_eigenvalue(rho, u, p)
        dt = dx*CFL/ymax
        if t + dt > 25:
            dt = 25 - t
            if verbose:
                print("Last Courant: %.2f" % (dt/dx*ymax))
        rho, u, p = scheme(rho, u, p, dt, dx, gamma=gamma, limiter=limiter)
        t += dt
```

```python
def format_ax(ax):
    ax[0].set_title('Density', size=20)
    ax[1].set_title('Velocity', size=20)
    ax[2].set_title('Pressure', size=20)
    ax[2].set_xlabel('Position')
    ax[2].set_xlim(-50, 50)
    ax[0].set_ylabel(r'$\rho$', size=17)
    ax[1].set_ylabel('$u$', size=17)
    ax[2].set_ylabel('$P$', size=17)
    ax[0].legend(loc='upper left', fancybox=True, framealpha=0.5, 
                 numpoints=1, )
    for axes in ax:
        format_axes(axes)
```

```python
def add_data(rho, u, p, style, label, ax):
    ax[0].plot(x, rho, style, label=label)
    ax[1].plot(x, u, style)
    ax[2].plot(x, p, style)

def add_exact(ax):
    add_data(rho_ex, u_ex, p_ex, style='k-', label='Exact', ax=ax)

def add_limiter(limiter, style, ax, N=5, verbose=False, CFL=None):
    if CFL is None:
        rho, u, p = integrate(limiter, rho0, u0, p0, verbose=verbose)
    else:
        rho, u, p = integrate(limiter, rho0, u0, p0, verbose=verbose, CFL=CFL)
    add_data(rho, u, p, style, label=limiter, ax=ax)
```

```python
def get_error(limiter):
    rho, u, p = integrate(limiter, rho0, u0, p0, CFL=0.1, verbose=False)
    err_rho = np.linalg.norm(rho - rho_ex)/(len(x)**(1/2))
    err_u = np.linalg.norm(u - u_ex)/(len(x)**(1/2))
    err_p = np.linalg.norm(p - p_ex)/(len(x)**(1/2))
    err_g = sum((err_rho, err_u, err_p))/3
    return err_g
```
