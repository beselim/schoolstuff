### Exercise 1-3. Determination of the global order of accuracy.

```python
import pdb
```

```python
def exercise1_3(signal, scheme, pdf, title):
    # Helper function to plot and calculate order of accuracy
    left = 0
    right = 400
    fig, ax = plt.subplots(figsize=(12, 8))
    courant = 0.8
    solver = solvers.Solver(scheme, signal)
    # List of delta_t to be used
    DT = np.array([0.5, 0.25, 0.125, 0.0625])
    # log(1/dx) for each delta_t above. 
    DX = []
    norms = []
    for dt in DT:
        solver.solve(courant, dt)
        dx = solver.dx
#         print(dx)
        DX.append(dx)
        x = solver.x
        # Only evaluate error on range around signal

#         cond = np.logical_and(x >= left, x <= right)
#         x = x[cond]
        # U at last time step (t = 100)
        u = solver.u[-1]
#         u = u[cond]
        _, u_exact = calc_exact(time=100, signal=signal, x=x)
        err = u - u_exact
#         norm = np.linalg.norm(err)/len(x)
        norm = np.linalg.norm(err)/(len(x)**(1/2))
        norms.append(norm)
    mlog_DT = -np.log(DT)
    log_norms = np.log(norms)
    mlog_DX = -np.log(DX)
    # Only use last two points for global order of accuracy
    DT_calc = mlog_DT[-2:]
    DX_calc = mlog_DX[-2:]
    norms_calc = log_norms[-2:]
#     pdb.set_trace()
    A = np.vstack([DT_calc, np.ones(len(DT_calc))]).T
    R, c = np.linalg.lstsq(A, norms_calc)[0]
    line = R*DT_calc+c # regression line
    ax.plot(DT_calc,line,'r-', label='R = %.2f' % R)
    ax.plot(mlog_DT, log_norms, '.', markersize=10)
    ax.set_xlabel(r'$\log\frac{1}{\Delta t}$', fontsize='x-large')
    ax.set_ylabel(r'$\log\Vert e\Vert$', fontsize='x-large')
    ax.legend(fancybox=True, bbox_to_anchor=(0.8, 0.8), fontsize='x-large')
#     ax.set_title(title)
    format_axes(ax)
    plt.tight_layout()
    fig.savefig('mini_project_2_tex/figs/' + pdf)
    plt.show()
```

```python
exercise1_3(triangle, schemes.lax_friedrichs, '1-3_LF_triangle.pdf', 'Lax-Friedrichs - Triangular Signal')
```

```python
exercise1_3(smooth, schemes.lax_friedrichs, '1-3_LF_smooth.pdf', 'Lax-Friedrichs - Smooth Signal')
```
