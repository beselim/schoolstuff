```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import os
from euler import schemes
from riemann.riemann import Riemann
from riemann.plot import plot_riemann, plot_particles
from linear.plot_utils import format_axes, latexify
%matplotlib inline
```

```python
latexify()
make_path = lambda x: os.path.abspath(os.path.join('mini_project_2_tex', 'figs', x + '.pdf'))
```

```python
# Initial conditions
rho_l = 1
u_l = 0
p_l = 2
rho_r = 1
u_r = 0 
p_r = 1
gamma = 1.4
t_end = 25
dx = 1
nx = 100
```

```python
# Generating arrays
x = np.arange(-49.5, 49.5 + dx, dx)
left = x < 0
right = x > 0
rho0 = np.ones_like(x)*rho_l
rho0[right] = rho_r
u0 = np.ones_like(x)*u_l
u0[right] = u_r
p0 = np.ones_like(x)*p_l
p0[right] = p_r
```

```python
# Primitive values
W_l = np.array([rho_l, u_l, p_l])
W_r = np.array([rho_r, u_r, p_r])
```

```python
# Find exact solution
riemann = Riemann(W_l, W_r, gamma)
riemann.update()
rho_ex, u_ex, p_ex = riemann.evaluate_states(x, t_end).T
```

### Exact solution

```python
fig, ax = plt.subplots(figsize=(10,6))
plot_riemann(riemann, np.arange(0, t_end + 0.1, 0.1), ax)
# ax.set_title('Riemann representation', size=20)
ax.set_xlabel('$x$', size=17)
ax.set_ylabel('$t$', size=17)
# ax.legend()
format_axes(ax)
fig.savefig(make_path('riemann'))
plt.show()
```

```python
def max_eigenvalue(rho, u, p):
    c = np.sqrt(gamma*p/rho)
    # Calculate maximum of each eigenvalue
    y1max = np.max(np.abs(u - c))
    y2max = np.max(np.abs(u))
    y3max = np.max(np.abs(u + c))
    ymax = max((y1max, y2max, y3max))
    return ymax
```

```python
def integrate(scheme, rho0, u0, p0, CFL, verbose=True):
    rho, u, p = rho0.copy(), u0.copy(), p0.copy()
    t = 0
    while True:
        if t >= 25:
            return rho, u, p
        ymax = max_eigenvalue(rho, u, p)
        dt = dx*CFL/ymax
        if t + dt > 25:
            dt = 25 - t
            if verbose:
                print("Last Courant: %.2f" % (dt/dx*ymax))
        rho, u, p = scheme(rho, u, p, dt, dx, gamma=gamma)
        t += dt
```

```python
def make_plot(name, rho, u, p):
    path = make_path(name)
    ratio = 11/8.5
    fig, ax = plt.subplots(nrows=3, sharex=True, figsize=(12, 12*ratio))
    ax[0].set_title('Density', size=20)
    ax[1].set_title('Velocity', size=20)
    ax[2].set_title('Pressure', size=20)
    ax[2].set_xlabel('Position')
    ax[2].set_xlim(-50, 50)
    ax[0].set_ylabel(r'$\rho$', size=17)
    ax[1].set_ylabel('$u$', size=17)
    ax[2].set_ylabel('$P$', size=17)

    ax[0].plot(x, rho_ex, 'r-', x, rho, 'b.')
    ax[1].plot(x, u_ex, 'r-', x, u, 'b.')
    ax[2].plot(x, p_ex, 'r-', x, p, 'b.')
    for axes in ax:
        format_axes(axes)
    plt.tight_layout()
    fig.savefig(path)
    plt.show()
```

### Godunov

```python
scheme = schemes.godunov
CFL = 0.9
rho_g, u_g, p_g = integrate(scheme, rho0, u0, p0, CFL=CFL)
make_plot('godunov', rho_g, u_g, p_g)
```

### Roe

```python
scheme = schemes.roe
CFL = 0.9
rho_r, u_r, p_r = integrate(scheme, rho0, u0, p0, CFL)
make_plot('roe', rho_r, u_r, p_r)
```

### McCormack

```python
CFL = 0.9
scheme = schemes.mccormack
rho_m, u_m, p_m = integrate(scheme, rho0, u0, p0, CFL)
make_plot('mccormack', rho_m, u_m, p_m)
```
