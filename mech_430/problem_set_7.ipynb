Here is the list of required equations for this problem:
\begin{align}
&\text{Normal Shock Relation}  &&\text{Oblique Shock Relations}\\\\
M_y &= \sqrt{\dfrac{2+(\gamma-1)M_x^2}{2\gamma M_x^2 - (\gamma -1)}}&\frac{P_y}{P_x} &= \dfrac{2\gamma M_x^2\sin^2\sigma - (\gamma - 1)}{\gamma + 1} \\
&  &\tan\delta &= \dfrac{(M_1^2\sin^2\sigma - 1)\cot\sigma}{\frac{\gamma + 1}{2}M_1^2-M^2_1\sin^2\sigma + 1}\\
\end{align}

<h4> First Shock </h4>

Using these relations, we can plot the *shock polar* for the first shock with the following process:
* Start from given: $M_1 = 4$.
* Pick a value for $\sigma_1$ in the range [$\alpha_1$,90] where $\alpha_1 = \sin^{-1}\frac{1}{M_1}$
* Compute the resulting value for $\delta_1$.
* Compute pressure ratio $P_2/P_1$
* Compute $M_2$

Numpy (Python package) is quite a fine tool and enables us to do the above for all required values of $\sigma_1$ in one shot, without any loops.

<h4> Second shock </h4>

The process for obtaining the *shock polar* for the second shock is similar to the above. However, plotting a shock polar requires a particular value of Mach Number $M$. That value depends on whether we're in case (b) or case (c). The process is the following:
* Use previously calculated $M_{2,(b)}$ corresponding to the deflection angle in (b)
* Pick a value for $\sigma_{2}$ in the range [$\alpha_2$,90] where $\alpha_2 = \sin^{-1}\frac{1}{M_2}$
* Compute the resulting value for $\delta_2$.
* Compute pressure ratio $P_3/P_1 = P_3P_2/P_2P_1$
* Compute $M_3$

The above must be repeated for (c) by picking the value of $M_2$ corresponding to the deflection angle in (c)

```python
#Translate above relations into functions
def normalShock(Mx):
    num = 2 + (gamma - 1)*Mx**2
    den = 2*gamma*Mx**2 - (gamma - 1)
    return np.sqrt(num/den)

# def get_P_ratio(Mx):
#     num = 2*gamma*Mx**2 - (gamma - 1)
#     den = gamma + 1
#     return num/den

def getPratio(Mx,sigma):
    num = 2*gamma*(Mx*np.sin(sigma.to('radians')))**2 - (gamma - 1)
    den = gamma + 1
    return num/den

def getDelta(Mx,sigma):
    sigma_rad = sigma.to('radians')
    num = ((Mx*np.sin(sigma_rad))**2 - 1)/np.tan(sigma_rad)
    den = (gamma+1.0)/2.0*Mx**2 - (Mx*np.sin(sigma_rad))**2 + 1
    return np.arctan(num/den)
```

```python
def findM(delta_array,delta_value,M_array):
    idx = (np.abs(delta_array-delta_value)).argmin()
    return M_array[idx]
```

```python
#Constant
gamma = 1.4
h = 100 #Step
```

```python
############
#Inlet flow
############
M1 = 4.0 
alpha1 = Q_(np.arcsin(1/M1),'radians')
sigma1 = Q_(np.linspace(alpha1.to('degrees').magnitude,90,h),'degrees')
# sigma1 = Q_(45.2,'degrees')
M1_N = M1*np.sin(sigma1.to('radians'))
```

```python
############
#First shock
############
P2_over_P1 = getPratio(M1,sigma1)
delta1 = getDelta(M1,sigma1).to('degrees')

M2_N = normalShock(M1_N)
M2_all = M2_N/np.sin(sigma1.to('radians')-delta1.to('radians'))
```

```python
#############
#Second shock 
#Delta = 20 (b)
#############
delta = Q_(20,'degrees')
M2_b = findM(delta1[:h/2],delta,M2_all)

alpha2_b = Q_(np.arcsin(1/M2),'radians')
# sigma2 = Q_(42,'degrees')
sigma2_b = Q_(np.linspace(alpha2_b.to('degrees').magnitude,90,h),'degrees')

P3_over_P2_b = getPratio(M2_b,sigma2_b)
delta2_b = getDelta(M2_b,sigma2_b).to('degrees')
```

```python
##############
#Second shock
#Delta = 30 (c)
##############
delta = Q_(30,'degrees')
M2_c = findM(delta1[:h/2],delta,M2_all)

alpha2_c = Q_(np.arcsin(1/M2_c),'radians')
# sigma2 = Q_(42,'degrees')
sigma2_c = Q_(np.linspace(alpha2_c.to('degrees').magnitude,90,h),'degrees')

P3_over_P2_c = getPratio(M2_c,sigma2_c)
delta2_c = getDelta(M2_c,sigma2_c).to('degrees')
```

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(delta1,sigma1,'k-')
ax.plot(delta2_b,sigma2_b,'b-')
ax.plot(delta2_c,sigma2_c,'g-')
ax.legend(['First','Second (b)','Second (c)'],loc='best', fancybox=True, framealpha=0.5)
ax.set_xlabel('Deflection angle'); ax.set_ylabel('Shock angle')
ax.set_title('Shock angle vs Deflection Angle')
ax.grid()
plt.show()
```
