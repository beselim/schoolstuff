```python
import numpy as np
import matplotlib.pyplot as plt
from riemann.riemann import Riemann
from riemann.plot import plot_riemann, plot_particles
from riemann import waves
%matplotlib inline
```

```python
# Sampling
N = 1000
x = np.array([-0.5 + (i - 0.5)/N for i in range(1, 1000 + 1)])
```

## Problem 1.
$W_L = [8.0, 0.0, 480.0]$

$W_R = [1.0, 0.0, 1.0]$

$x_0 = 0$

$\gamma = 5/3$

$t_1 = 0.04$

```python
W_l = np.array([8.0, 0.0, 480.0])
W_r = np.array([1.0, 0.0, 1.0])
x0 = 0
gamma = 5.0/3.0
t1 = 0.04
riemann1 = Riemann(W_l, W_r, gamma, x0)
W1 = riemann1.evaluate_states(x, t1)
```

## Problem 2.
$W_L = [1.0,-2.0,0.4]$

$W_R = [1.0,2.0,0.4]$

$x_0 = 0$

$\gamma = 1.4$

$t_1 = 0.15$

```python
W_l = np.array([1.0, -2.0, 0.4])
W_r = np.array([1.0, 2.0, 0.4])
x0 = 0
gamma = 1.4
t2 = 0.15
riemann2 = Riemann(W_l, W_r, gamma, x0)
W2 = riemann2.evaluate_states(x, t2)
```

## Problem 3.
$W_L = [1.0,−19.59745,1000.0]$

$W_R = [1.0,-19.59745,0.01]$

$x_0 = 0.3$

$\gamma = 1.4$

$t_1 = 0.012$

```python
W_l = np.array([1.0, -19.59745, 1000.0])
W_r = np.array([1.0, -19.59745, 0.01])
x0 = 0.3
gamma = 1.4
t3 = 0.012
riemann3 = Riemann(W_l, W_r, gamma, x0)
W3 = riemann3.evaluate_states(x, t3)
```

```python
def plot_test(riemann, W, t, i, axes):
    x0 = riemann.x0
    plot_riemann(riemann, np.linspace(0,t*1.2, 100), axes[0])
    axes[0].plot(x, np.ones_like(x)*t, 'g-.')
    axes[0].set_xlim(-0.5, 0.5)
    axes[0].set_title('Problem %i' % i, size=30)
    titles = ['Density', 'Velocity', 'Pressure']
    for i in range(3):
        axes[i+1].plot(x, W[:,i], 'b-')
        axes[i+1].set_xlim(-0.5, 0.5)
        axes[i+1].set_title(titles[i], size = 20)
    intersect = []
    for wave in [riemann.left_wave, riemann.right_wave]:
        try:
            intersect.append(wave.speed[0]*t + riemann.x0)
            intersect.append(wave.speed[1]*t + riemann.x0)
        except IndexError:
            intersect.append(wave.speed*t + riemann.x0)
    intersect.append(riemann.u_star*t + riemann.x0)
    for ax in axes:
        for point in intersect:
            ax.axvline(point, color='black', linestyle='-.')
```

```python
fig, axes = plt.subplots(nrows=4, ncols=3, sharex=True, squeeze=True, figsize=(30, 30))
axes = np.array(axes).T
axes1, axes2, axes3 = axes
plot_test(riemann1, W1, t1, 1, axes1)
plot_test(riemann2, W2, t2, 2, axes2)
plot_test(riemann3, W3, t3, 3, axes3)
plt.tight_layout()
fig.savefig('project1/tex/figs/plots.pdf')
```

```python
def write_results(W, target_file):
    np.savetxt(target_file, np.hstack((np.reshape(x, (1000, 1)), W)), fmt = '%12.5e', delimiter = ' ', 
               header='%12s %12s %12s %12s' % ('x', 'rho', 'u', 'p'), comments='')
```

```python
write_results(W1, 'project1/ascii_results/problem1.txt')
write_results(W2, 'project1/ascii_results/problem2.txt')
write_results(W3, 'project1/ascii_results/problem3.txt')
```
