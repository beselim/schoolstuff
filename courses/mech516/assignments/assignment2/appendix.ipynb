### Problem 3 b)

```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi
%matplotlib inline
```

```python
t = np.linspace(0, 2*pi, 501)
```

```python
def ax_with_circle():
    fig, ax = plt.subplots(figsize=((8,8)))
    unit_circle = plt.Circle(xy=(0,0), radius=1, facecolor='k', edgecolor='none', alpha=0.1, label='Stability Region')
    ax.add_artist(unit_circle)
    ax.set_xlim(-2,2)
    ax.set_ylim(-2,2)
    return ax
```

```python
all_s = [0.25, 0.5, 0.75, 1.0, 1.1]
styles = ['b-', 'r-', 'g-', 'b--', 'r--']
```

```python
def e_re(coeff):
    return np.cos(coeff*t)
def e_im(coeff):
    return np.sin(coeff*t)
```

```python
ax = ax_with_circle()
for s, style in zip(all_s, styles):
    re = 1 + s/2*(e_re(-1) - e_re(1)) + s**2/2*(e_re(-1) + e_re(1) - 2)
    im = s/2*(e_im(-1) - e_im(1)) + s**2/2*(e_im(-1) + e_im(1))
    ax.plot(re, im, style, label='s=%.2f'%s)
ax.legend()
plt.show()
```

### Problem 3 c)

```python
all_s = [0.26, 0.28, 0.30, 0.32, 0.34]
ax = ax_with_circle()
for s, style in zip(all_s, styles):
    re = 1 - s/2*(3 - 4*e_re(-1.0) + e_re(-2.0)) + s**2/2*(1 - 2*e_re(-1.0) + e_re(-2.0))
    im = s/2*(-4*e_im(-1.0) + e_im(-2.0)) + s**2/2*(-2*e_im(-1.0) + e_im(-2.0))
    ax.plot(re, im, style, label='s=%.2f'%s)
ax.legend(loc='best', fancybox=True, framealpha=0.5)
ax.set_xlim(-1,1,1.1)
ax.set_ylim(-1.1,1.1)
plt.show()
```
