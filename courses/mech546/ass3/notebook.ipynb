```python
import sympy as sp  # Symbolic Math
import numpy as np  # Linear Algebra
import matplotlib.pyplot as plt  # Plotting
sp.init_printing()
%matplotlib inline
```

# Problem 1

### Governing Differential Equation

$$
-\frac{\text{d}}{\text{d}x}\left[ (1 + x) \frac{\text{d}u}{\text{d}x} \right] = 0 \quad \text{for} \quad 0 < x < 1 \\
u(0) = 0, ~~ u(1) = 1
$$

### Ritz Approximation
$$
u(x) \approx U_2 = \phi_0(x) + \sum_{i=1}^{2} \phi_i(x)c_i
$$
where
$$
\phi_0 = x, ~~ \phi_i = x^i(1 - x)
$$
The main goal is to find the coefficients $c_i$.
### Analytical Solution

$$
u(x) = \frac{\log(x + 1)}{\log(2)}
$$

```python
# Load custom plotting config
execfile('../../matplotlibrc.py')
```

```python
x = sp.symbols('x')
```

```python
def phi(idx):
    return x**idx * (1 - x)

def dphi(idx):
    return idx*x**(idx - 1) - (idx + 1)*x**idx

def Kij(i,j):
    integrand = (1+x)*dphi(i)*dphi(j)
    return sp.integrate(integrand, (x, 0, 1))

def Fi(i):
    integrand = (1+x)*dphi(i)*1
    return sp.integrate(integrand, (x, 0, 1))
```

```python
# Construct K and F
K = sp.zeros(2,2)
F = sp.zeros(2,1)
for i in range(2):
    F[i] = -Fi(i+1)
    for j in range(2):
        K[i,j] = Kij(i+1,j+1)
print("K = %s" % K)
print("F = %s" % F)
```

```python
# Solve for C
C = K.inv().dot(F)
print("C = %s" % C)
```

```python
# Construct U
u = sum([C[i]*phi(i+1) for i in range(2)]) + x
u.simplify()
```

```python
X = np.linspace(0, 1)

# Approximate Solution
uritz = sp.lambdify(x,u)(X)

# Exact Solution
ux = np.vectorize(sp.lambdify( x, sp.log(x + 1.)/sp.log(2.) ))(X)
```

```python
# Calculate Error
err = 100*(uritz[1:] - ux[1:])/ux[1:]
maxerr = np.abs(err).max()
```

```python
# Add tick to largest value
def tick_max(y, ax):
    idx_max = np.abs(y).argmax()
    val = y[idx_max]
    ticks = list(ax.get_yticks())
    ticks.append(val)
    ax.set_yticks(ticks)
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, sharex=True, figsize=(12,14))
ax1.plot(X, uritz, 'o', label='Ritz', ms=4)
ax1.plot(X, ux, label='Analytical')
ax1.legend(numpoints=1)
ax1.set_ylabel('Displacement $u(x)$')
ax1.set_title('Comparison of two-parameter Ritz solution and analytical solution.')
ax2.set_ylabel('Relative Error (%)')
ax2.set_xlabel('Position $x$')
ax2.plot(X[1:], err)
ax2.plot((0, 1), (0, 0), 'k--')
ax2.set_ylim(-maxerr*1.05, maxerr*1.05)
tick_max(err, ax2)
fig.tight_layout()
plt.show()
```

The Ritz approximation yields a maximum relative error of about 1.4 %.

# Problem 2

### Governing Differential Equation

$$
-2u\frac{\text{d}^2u}{\text{d}x^2} + \left( \frac{\text{d}u}{\text{d}x} \right)^2 = 4 \quad 0 < x < 1 \\
u(0) = 1, ~~ u(1) = 0
$$

### Ritz Approximation
$$
U_1(c_1, x) = c_1 x (1 - x) + (1 - x)
$$

### Analytical Solution

$$
u_0(x) = 1 - x^2
$$

```python
def U1(x, c1):
    return c1*x*(1 - x) + (1 - x)
    
def u0(x):
    return 1 - x**2
```

```python
C = [-3, -1, 1]
x = np.linspace(0, 1)
```

```python
fig, ax = plt.subplots(figsize=(14, 10))

for c1 in C:
    ax.plot(x, U1(x, c1), label='$c_1$ = %i' % c1)
ax.set_xlabel('Position $x$')
ax.set_ylabel('Displacement $u$')
ax.set_title('Comparison of one-parameter Ritz approximation and analytical solution.\n'
             '$c_1$ = 1 is the exact solution.')
ax.legend()
plt.show()
```

It is interesting to see the difference in potential solutions obtained by the Ritz approximations!
