## Governing Differential Equation

$$
\frac{d^2 u}{dx^2} + 2u - 1 = 0 
$$

with boundary conditions:

$$
u(0) = 0 \quad u(1) = 0
$$

## Analytical Solution

The ODE above can be solved using MATLAB's symbolic math toolbox with the 
following code:
```python
syms u(x);
dsolve(diff(u, 2) + 2*u - 1 == 0, u(0) == 0, u(1) == 0)
```

which gives the exact analytical solution:
$$
u(x) = \frac{\sin{\left (\omega x \right )}}{2 \sin{\left (\omega \right )}} \left(\cos{\left (\omega \right )} - 1\right) - \frac{1}{2} \cos{\left (\omega x \right )} + 0.5\\
\omega = \sqrt{2}
$$
