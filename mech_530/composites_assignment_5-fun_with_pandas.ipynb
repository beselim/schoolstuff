```python
#Create a Simulation object composed of a laminate.
sim = Sim(laminate = Laminate('0_2/p25/0_2s',
                               materialID = 5, #5
                               core_thick = 0.01))
```

```python
#Define and apply load
P = Q_(1000,'N'); b = Q_(0.11,'m'); L = Q_(0.51,'m')
M1 = -P*L/(4*b); N1 = 0.5*P/b;
M = Q_([M1.magnitude,0,0],M1.units)
N = Q_([N1.magnitude,0,0],N1.units)
# print M
# print N
```

**Loads**

```python
print "M [N] : "
print [round(M1.magnitude,2),0,0]
print "\nN [N/m] : "
print [round(N1.magnitude,2),0,0]
```

```python
#Apply load
sim.apply_M(M)
sim.apply_N(N)
sim.solve()
```

**Maximum stress failure criteria**

```python
def find_min(dataframe,columns):
    nonzero_dat = dataframe[dataframe != 0]
    for col in columns:
        the_data = nonzero_dat[[col,'Ply']]
        min_index = the_data[col].idxmin()
        print the_data.iloc[min_index].to_string(name=True)
        print '----'
```

```python
find_min(R_data,['FT_max','FC_max','MT_max','MC_max','S'])
```

```python
R_max = R_data[['FT_max','FC_max','MT_max','MC_max']][R_data != 0]
tempframe = R_max.min().to_frame(name = 'Lowest R')
tempframe
```

```python
tempframe['Ply'] = R_data['Ply'].iloc[R_max.idxmin()].values
```
